```python
import time
import json
import requests
from bs4 import BeautifulSoup
import pandas as pd

comment = []
star = []
result = {}

for page in range(1,10):
    base_url = 'https://store-kr.uniqlo.com/displayDetail/listGoodsAssessment.lecs?nolayout=yes&storeNo=83&siteNo=50706&goodsNo=NQ31112670&orderType=1&currentPage={}&itemCount=842&perPage=10&perPageNavi=10'
    my_headers = {
        'referer' : 'https://store-kr.uniqlo.com/display/showDisplayCache.lecs?goodsNo=NQ31112670&infw_disp_no_sct_cd=10&stonType=P&storeNo=83&siteNo=50706',
        'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.79 Safari/537.36'
    }
    
    response = requests.get(base_url.format(page), headers = my_headers)
    soup = BeautifulSoup(response.content, "lxml")
    
    match=soup.select('td.txt07 > div.reTxt')
    match1=soup.select('td.txt02 span.blind')
    for index in range(len(match)):
        comment.append(match[index].text.strip())
        star.append(match1[index].text.strip())

result = {'별점' : star, '상품평' : comment}
df = pd.DataFrame(result)
```
